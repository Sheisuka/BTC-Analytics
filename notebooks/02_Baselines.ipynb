# 1. Базовая модель, основанная лишь на значениях MA (скользящее среднее)
В качестве целевых признаков возьмем up_d_ahead. Модель строится по следующему правилу: *если значение btc_close превышает значение btc_ma, тогда предсказанное значение целевого признака равно единице*.  
Рассмотрим всевозможные комбинации интервала скользящей средней и длины горизонта предсказания. Зафиксируем лучшую комбинацию в качестве базовой модели

```python
def ma_predict(df, ma_period, horizon):
    ma_col = f"btc_ma{ma_period}"
    y_col  = f"up_{horizon}d_ahead"

    assert df["btc_close"].notna().all()
    assert df[ma_col].notna().all()
    assert df[y_col].notna().all()

    y      = df[y_col]
    y_pred = df["btc_close"] > df[ma_col]

    TP = ((y == 1) & (y_pred == 1)).sum()
    TN = ((y == 0) & (y_pred == 0)).sum()
    FP = ((y == 0) & (y_pred == 1)).sum()
    FN = ((y == 1) & (y_pred == 0)).sum()

    acc = (TP + TN) / (TP + TN + FP + FN)
    prec = TP / (TP + FP)
    rec = TP / (TP + FN)

    res = {
        "TP": TP,
        "TN": TN,
        "FP": FP,
        "FN": FN,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
    }

    return res
```

```python
results_ma = []
for period in (7, 20, 60):
    for hor in (1, 7, 30):
        res = ma_predict(dataset, period, hor)
        res["period"]  = period
        res["horizon"] = hor
        results_ma.append(res)

res_ma = pd.DataFrame(results_ma)

display(res_ma)
```

# 2. Базовая модель, всегда вверх
В качестве целевых признаков возьмем up_d_ahead. Модель строится по правилу *всегда вверх*
Рассмотрим всевозможные длины горизонта предсказания. Зафиксируем лучшую в качестве базовой модели

```python
def always_up_predict(df, horizon):
    y_col  = f"up_{horizon}d_ahead"

    assert df["btc_close"].notna().all()
    assert df[y_col].notna().all()

    y      = df[y_col]
    y_pred = pd.Series(1, index=y.index)

    TP = ((y == 1) & (y_pred == 1)).sum()
    TN = ((y == 0) & (y_pred == 0)).sum()
    FP = ((y == 0) & (y_pred == 1)).sum()
    FN = ((y == 1) & (y_pred == 0)).sum()

    acc = (TP + TN) / (TP + TN + FP + FN)
    prec = TP / (TP + FP)
    rec = TP / (TP + FN)

    res = {
        "TP": TP,
        "TN": TN,
        "FP": FP,
        "FN": FN,
        "accuracy": acc,
        "precision": prec,
        "recall": rec,
    }

    return res
```

```python
results_always_up = []
for hor in (1, 7, 30):
    res = always_up_predict(dataset, hor)
    res["horizon"] = hor
    results_always_up.append(res)

res_always_up = pd.DataFrame(results_always_up)

display(res_always_up)
```
